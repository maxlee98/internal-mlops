```python
start_year = 2005
end_year = 2009
```

```python

# Loop over the years from 2000 to 2005
for year in range(end_year, start_year-1, -1):
    # Loop over the months from January to December
    for month in range(1, 13):
        # Append the date to the list
        for _ in range(4):
            print(f"{month}/01/{year}")
```
