```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

current_dir = os.getcwd()
data_fldr = os.path.join(current_dir, "data")
```

```python
residential_df = pd.read_excel(os.path.join(data_fldr, "ConsumerPriceIndex.xlsx"), sheet_name="Sheet16")
residential_df['Date'] = residential_df['Date'].astype(str)
residential_df
```

```python
def convert_to_date(year_quarter):
    year, quarter = str(year_quarter).split(' ')
    # print(year, quarter)
    month = int(quarter[0]) * 3 - 2  # Convert quarter to month (e.g., 1Q -> January, 2Q -> April, etc.)
    date_str = f'01-{month:02d}-{year}'  # Construct date string in "dd-mm-yyyy" format
    return pd.to_datetime(date_str, format='%d-%m-%Y').strftime('%d-%m-%Y')

residential_df['Date'] = residential_df['Date'].apply(convert_to_date)
```

```python
residential_df['Date'] = pd.to_datetime(residential_df['Date'], format='%d-%m-%Y')
residential_df['Year'] = residential_df['Date'].dt.year
residential_df
```

```python
residential_df_groupby = residential_df.groupby(['Year']).mean().reset_index()
```

```python
residential_df_groupby.to_csv("residential_df_groupby.csv", index=False)
```
